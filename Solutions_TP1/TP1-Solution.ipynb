```python
V_titrage=np.array([]) #volume versé en ml
conductimetrie=np.array([]) #grandeur mesuré pour le titrage

Vol_derivee=np.array([]) #nouvel liste pour la dérivée
conduct_derivee=np.array([]) #liste dérivée

for i in range(1,len(V_titrage)-1):
    conduct_derivee.append((conduct_derivee[i+1]-conduct_derivee[i-1])/(Vol_derivee[i+1]-Vol_derivee[i-1])) #On calcule la dérivée en prenant le point après et le point avant et on calcule la pente
    Vol_derivee.append(V_titrage[i])

max_derivee=np.max(conduct_derivee)
max_index=conduct_derivee.index(max_derivee) #trouve l'indice pour le max de la dérivée
V_max=Vol_derivee[max_index] #trouve le volume pour laquelle la dérivée est max
print(V_max) 

plt.plot(V_titrage,conductimetrie)
plt.scatter(V_titrage,conductimetrie,label="Points expérimentaux") #trace les points expérimentaux

plt.plot(Vol_derivee,conduct_derivee,c='r',label=f"Courbe de la dérivée \n V_eq={V_max}") #trace la dérivée et affiche le volume équivalent

plt.ylabel('conductimétrie')
plt.xlabel('Volume versée (en mL)')
plt.grid()
plt.legend()
#plt.savefig("Titrage_2.eps",format='eps') #sauvegarde la figure, on peut mettre .png au lieu de .eps et format="png"
plt.show()

```
