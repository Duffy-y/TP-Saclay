```python
"""
    Caméra linéaire
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("a=70_d=300.csv", delimiter=";")

pixel = df.iloc[:,0].values
intensity = df.iloc[:,1].values

pixel *= 14 * 10**(-4)
a1 = 2.17
a2 = 0.94

a3 = 1.46
a4 = 1.64

plt.plot(pixel, intensity, color="red")
plt.ylabel("Intensité lumineuse reçue")
plt.xlabel("Position du pixel (en cm)")
plt.axvline(a1, ymin = 0, ymax=0.2)
plt.axvline(a2, ymin = 0, ymax=0.2)
plt.text(a1 - 0.05, 0.2, f"x={a1} +/- 0.02 cm", color="b")
plt.text(a2, 0.2, f"x={a2} +/- 0.02 cm", horizontalalignment='right', color="b")

plt.axvline(a3, ymin = 0, ymax=1, color="green")
plt.axvline(a4, ymin = 0, ymax=1, color="green")
plt.text(a3 - 0.4, 0.9, f"x={a3} +/- 0.02 cm", horizontalalignment='right', color="green")
plt.text(a4 + 0.3, 0.9, f"x={a4} +/- 0.02 cm", color="green")
plt.savefig("graphe_camera_lineaire.eps")
```
